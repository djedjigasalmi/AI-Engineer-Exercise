```python
import numpy as np
import os
import PIL
from PIL import Image###
import tensorflow as tf
import pandas as pd
#!pip install tensorflow_datasets
import tensorflow_datasets as tfds
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
import matplotlib.pyplot as plt
import cv2
```

```python
print(tf.__version__)
```

```python
# Set the path to the folder containing the images
data_dir = "dataset"

# Load the dataset using TensorFlow
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    batch_size=32,
    image_size=(224, 224)
)

# Get a list of the class names from the dataset
class_names = dataset.class_names

# Get the number of classes and images
num_classes = len(class_names)
num_images = len(dataset.file_paths)

# Iterate through the classes and count the number of images in each class
for class_name in class_names:
    class_path = os.path.join(data_dir, class_name)
    num_images = len(os.listdir(class_path))
    print("{}: {} images".format(class_name, num_images)) 
```

```python
# Create a list of image paths for each class

fields_dir = os.path.join(data_dir, 'fields')
roads_dir = os.path.join(data_dir, 'roads')

fields_paths = [os.path.join(fields_dir, f) for f in os.listdir(fields_dir) if f.endswith('.jpg') | f.endswith('.jpeg')]
roads_paths = [os.path.join(roads_dir, f) for f in os.listdir(roads_dir) if f.endswith('.jpg') | f.endswith('.jpeg')]



# Concatenate the two lists to have an overall list of paths

all_paths = fields_paths + roads_paths

# Create the label images (0 for fields, 1 for roads)

labels = [0]*len(fields_paths) + [1]*len(roads_paths)

# Divide the date into training and validation data

train_paths, val_paths, train_labels, val_labels = train_test_split(all_paths, labels, test_size=0.2, random_state=42)

# Print the number of images in each set

print(f"Nombre d'images d'entraînement : {len(train_paths)}")
print(f"Nombre d'images de validation : {len(val_paths)}")
```

```python
# Create image generators for trainingand validation by using ImageDataGenerator to augment the data. 
# This helps to prvent overfitting and improve the eneralisation of the model. We will also normalize the images by dividing by 255.

# The flow_from_dataframe() method expects the label column to contain strings for binary classification

train_labels = np.array(train_labels).astype(str)
val_labels = np.array(val_labels).astype(str)

train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=20,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   horizontal_flip=True)

val_datagen = ImageDataGenerator(rescale=1./255)



# create two dataframes containing image paths and corresponding labels for training sets and validation 

train_df = pd.DataFrame({'path': train_paths, 'label': train_labels})
val_df = pd.DataFrame({'path': val_paths, 'label': val_labels})

# The flow_from_dataframe() method expects the label column to contain strings for binary classification

train_labels = np.array(train_labels).astype(str)
val_labels = np.array(val_labels).astype(str)

train_generator = train_datagen.flow_from_dataframe(dataframe=train_df,
                                                    x_col="path",
                                                    y_col="label",
                                                    batch_size=32,
                                                    shuffle=True,
                                                    class_mode="binary",
                                                    target_size=(224, 224))

val_generator = val_datagen.flow_from_dataframe(dataframe=val_df,
                                                x_col="path",
                                                y_col="label",
                                                batch_size=32,
                                                shuffle=False,
                                                class_mode="binary",
                                                target_size=(224, 224))
```

```python

# Create the CNN model
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D((2, 2)))
#model.add(Dropout(0.2))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
#model.add(Dropout(0.2)) 

#model.add(Conv2D(128, (3, 3), activation='relu'))
#model.add(MaxPooling2D((2, 2)))
#model.add(Dropout(0.2)) 

model.add(Flatten())
model.add(Dense(64, activation='relu')) 
model.add(Dense(1, activation='sigmoid'))

# Use Adam optimizer and the learning rate
from keras.optimizers import Adam
adam = Adam(learning_rate=0.0001)

# Compilate the model
model.compile(loss='binary_crossentropy', optimizer= adam, metrics=['accuracy'])

# Train the model on traing data and evaluate it on validation data
history = model.fit(train_generator, epochs=10, validation_data=val_generator)
```

```python
# Plot the accuracy as a function of epochs



plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')


test_loss, test_acc = model.evaluate(val_generator, verbose=2) 
```

```python
# Plot the loss as a function of number epochs

plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.ylim([0, 1])
plt.legend(loc='lower right')
test_loss, test_acc = model.evaluate(val_generator, verbose=2) 
```
